```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as sts
import math
```

```python
# Беру случайный параметр m - значение вокруг которого и будет строиться распределение, оно же матожидание в обеих функциях
m = 5.0
# Генерирую выборку объёма 1000 из Гамма распределения
myarray = sts.gamma.rvs(m,size=1000)
# Строю гистограмму выборки 
plt.hist(myarray, normed=True, bins=40, alpha=0.5, label='gamma variate - rvs')
plt.legend(loc='best') # задаю местоположение легенды
plt.ylabel('samples count') #подписываю оси
plt.xlabel('x')
```

# теоретическая плотность распределения случайной величины

```python
x=np.linspace(0,15,101) # задаю последовательность на промежутке от 0 до 15, в размере 101 точки - 100 промежутков
plt.plot(x, sts.gamma.pdf(x, m), 'r-', lw=3, alpha=0.7, label='gamma variate - pdf') # строю график
plt.legend(loc='best') # задаю местоположение легенды
```

# функция построения гистограммы распределений и плотности нормального распределения 

```python
def myfunction(rangesize, E, D):
    n = rangesize # объем выборки - подается на вход функции
    # гистограмма Гамма распределения
    raspred = np.array([sts.gamma.rvs(m, size=n) for x in xrange(1000)]) # генерирую выборки
    srednee = raspred.mean(axis = 1) # вычисляю выборочные средние
    plt.hist(srednee, normed=True, alpha=0.5, label='hist mean n = ' + str(n)) # строю гистограмму
        
    # математическое ожидание подается на вход функии
    mu = E
    # сигма нормального распределения вычисляется из последнего параметра, поданного на вход функции
    sigma = math.sqrt(D/n)
    print 'Для НОРМАЛЬНОГО распределения'
    print 'математическое ожидание =' , mu
    print 'сигма =' , sigma
    normalnoe = sts.norm(loc=mu, scale=sigma) # задаю распределение
    x = np.linspace(0,15,101) # задаю последовательность на промежутке от 0 до 15, в размере 101 точки
    pdf = normalnoe.pdf(x) # получаю массив значений нормального распределения в точках х
    plt.plot(x, pdf, 'r-', lw=3, alpha=0.7, label='normal pdf n = ' + str(n)) # строю график
    plt.ylabel('samples count') # подписываю оси
    plt.xlabel('x')
    plt.legend(loc='best') # задаю местоположение легенды
```

```python
#Вычисление теоритических параметров Гамма распределения:
E = sts.gamma.mean(m) # математического ожидания  
sq = sts.gamma.std(m) # среднеквадратического отклонения
D = sq**2             # дисперсии
myfunction(5, E, D)   # Запуск функции с вычисленными параметрами
print 'Для ГАММА распределения'
print 'Математическое ожидание =', E
print 'Дисперсия =', D
print 'Cреднеквадратическое отклонение =', sq
```

```python
myfunction(10, E, D) # Запуск функции с вычисленными параметрами и n = 10
```

```python
myfunction(30, E, D) # Запуск функции с вычисленными параметрами и n = 30
```

# Вывод

Увеличение выборки n (rangesize в функции myfunction) приводит к увеличению точности аппроксимации.
